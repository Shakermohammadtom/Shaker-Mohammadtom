```python
  # prompt user to enter name, number, type and email
    name = input("\nEnter contact name: ")
    phone = input("Enter contact phone number: ")
    type = input("Enter contact type: ")
    email = input("Enter contact email: ")
    
    # add attributes to the sublist
    list.append(name)
    list.append(phone)
    list.append(type)
    list.append(email)
    
    # add sublist to 2d list
    Phone_list.append(list)
    
    # inform user
    print("\nRecord successfully added!")

# this method searches a contact in the Phone_list    
def searchContact():
    
    # enter name
    name = input("\nEnter contact name to search: ")
    
    # loop to iterate over Phone_list
    flag = 0
    for list in Phone_list:
        
        # check for name
        if list[0] == name:
            flag += 1
            
            # if found then print below details 
            print("\nRecord found!")
            print("Name = " + list[0])
            print("Phone number = " + list[1])
            print("Phone type = " + list[2])
            print("Email = " + list[3])
    
    # if not found then print below message
    if flag == 0:
        print("\nRecord not found!")
        
# Main Function for Menu-Driven
def main():
    
    # loop until user wish to exit
    while True:
        
        # print MENU
        print("\nMENU:")
        print("1. Search a Contact")
        print("2. Add a new Contact")
        print("3. Delete a Contact")
        print("4. Update a Contact")
        print("5. Display all Contacts")
        print("6. Exit")
        
        # prompt for choice
        choice = int(input("\nEnter your choice: "))
        
        # check for choice and then call suitable method to perform next steps
        if choice == 1:
            searchContact()
        elif choice == 2:
            addContact()
        elif choice == 3:
            deleteContact()
        elif choice == 4:
            updateContact()
        elif choice == 5:
            displayContact()
        elif choice == 6:
            print("\nGood-bye!\n")
            break
        else:
            print("Invalid choice!")

# program starts from here
if __name__ == "__main__":
    
    # calling main
	print("### Welcome to Phone List ###")
	main()
```
